```python
import pandas as pd
```

```python
data = pd.read_csv('get_data/excelData/Results.csv')
```

```python
area = []
x = []
y = []
xm = []
ym = []
perim = []
bx = []
by = []
width = []
height = []
```

```python
i = 0
while(len(data) > 0):
    area.append(data['Area'][i])
    x.append(data['X'][i])
    y.append(data['Y'][i])
    xm.append(data['XM'][i])
    ym.append(data['YM'][i])
    perim.append(data['Perim.'][i])
    bx.append(data['BX'][i])
    by.append(data['BY'][i])
    width.append(data['Width'][i])
    height.append(data['Height'][i])
    data.drop([i],axis=0)
    i = i + 1
    if i == 100:
        break
```
