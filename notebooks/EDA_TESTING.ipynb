# `Data Information`

`Area`: The area of the object or shape.

`X, Y`: Coordinates of the center of mass of the object.

`XM, YM`: Coordinates of the centroid of the object.

`Perimeter`: The perimeter or circumference of the object.

`BX, BY`: Coordinates of the bounding box (or bounding rectangle) that encloses the object.

`Width`, Height: Dimensions of the bounding box (width and height).

`Result`: Possibly an outcome or a specific value related to the object.

`Type`: Could represent the type or category of the object.

# `PLOTTING GRAPHS` 

```python
plot = pd.read_excel('ALL_FEATURES_CLEANED_DATA.xlsx')
```

##### `PLOTTING NUMERICAL FEATURES`

```python
numerical_features = plot.columns[plot.dtypes != 'O']
```

```python
plot[numerical_features].head()
```

```python
for i in numerical_features:
    plt.figure(figsize=(7,3))
    sns.boxplot(plot[[i]])
    print(f'Maximum {i} of Seed --->',plot[i].max(),'mm²')
    print(f'Minimum {i} of Seed --->',plot[i].min(),'mm²')
```

## `HISTPLOTS`

```python
for i in numerical_features:
    plt.figure(figsize=(7,3))
    sns.histplot(plot[[i]],kde=True)
```

##### `PLOTTING CATEGORICAL FEATURES`

```python
categorical_features = plot.columns[plot.dtypes == 'O']
```

```python
plot[categorical_features].head()
```

```python
for i in categorical_features:
    plt.figure(figsize=(7,3))
    sns.histplot(plot[i],color='grey')
```

## PLOTTING INSIGHTS

```python
yes_data = plot[plot['Result'] == 'yes']
```

```python
no_data = plot[plot['Result'] == 'no']
```

## AREA of seed grown and not grown

```python
sns.histplot(yes_data[['Area']])
```

```python
sns.histplot(no_data[['Area']])
```

##### `CONCLUSION` : **Area of Seed between 75mm² to 85mm² has Highest Probability to Grow** and **Area of Seed near 90mm² has Least Probability to Grow** 

```python
plot["Type"].value_counts().plot.pie(y=plot["Type"],figsize = (5,5),autopct='%1.1f%%')
```
